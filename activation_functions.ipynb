```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
def ReLU(x):
    return max(0, x)
def tnh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
def leaky_relu(x):
    return max(0.1*x, x)
def linear_piecewise(x):
    if(x >= 0.5):
        return 1
    elif(x <= -0.5):
        return 0
    else:
        return x
def softmax(x):
    y = []
    sum = 0
    for i in x:
        sum+= math.exp(i)
    for i in x:
        y.append(math.exp(i)/sum)
    return y
```

```python
x = []
sigmoid_f = []
ReLU_f = []
tnh_f = []
leaky_ReLU = []
Linear_piecewise = []
x = np.arange(-10,10,0.5)
for i in x:
    sigmoid_f.append(sigmoid(i))
    ReLU_f.append(ReLU(i))   
    tnh_f.append(tnh(i))
    leaky_ReLU.append(leaky_relu(i))
    Linear_piecewise.append(linear_piecewise(i))
```

```python
plt.plot(x, sigmoid_f)
plt.show()
```

```python
plt.plot(x, ReLU_f)
plt.show()
```

```python
plt.plot(x, tnh_f)
plt.show()
```

```python
plt.plot(x, leaky_ReLU)
plt.show()
```

```python
plt.plot(x, Linear_piecewise)
plt.show()
```

```python
plt.plot(x, softmax(x))
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, sigmoid_f)
plt.title('Sigmoid')
plt.subplot(2, 3, 2)
plt.plot(x, ReLU_f)
plt.title('ReLU')
plt.subplot(2, 3, 3)
plt.plot(x, tnh_f)
plt.title('Tanh')
plt.subplot(2, 3, 4)
plt.plot(x, leaky_ReLU)
plt.title('Leaky ReLU')
plt.subplot(2, 3, 5)
plt.plot(x, Linear_piecewise)
plt.title('Linear Piecewise')
plt.subplot(2, 3, 6)
plt.plot(x, softmax(x))
plt.title('Softmax')

plt.tight_layout()
```
